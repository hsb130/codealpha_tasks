# Implementing a basic matrix factorization-based recommendation system using Singular Value Decomposition (SVD) in Python with the surprise library :


Of course! Let's break down the concept of a matrix factorization-based recommendation system using Singular Value Decomposition (SVD) in an easy-to-understand manner:

1. **What is a Recommendation System?**
   A recommendation system is like a virtual shopping assistant that suggests items (movies, music, products, etc.) to users based on their preferences and behaviors.

2. **Matrix Factorization:**
   - Imagine we have a large table (matrix) where rows represent users and columns represent items (e.g., movies, music tracks).
   - Each cell in the table represents a user's rating (or interaction) with an item.
   - Matrix factorization is a technique to decompose (break down) this large matrix into smaller matrices.
   - The goal is to find hidden patterns or features (latent factors) that explain the relationships between users and items.

3. **Singular Value Decomposition (SVD):**
   - SVD is a specific method for matrix factorization.
   - It decomposes the original matrix into three smaller matrices: U, Σ, and V^T.
   - U represents the relationship between users and latent factors.
   - Σ is a diagonal matrix containing singular values that indicate the importance of each latent factor.
   - V^T represents the relationship between items and latent factors.
   - By multiplying these matrices back together, we can approximate the original matrix and fill in missing values (unrated items).

4. **How SVD Helps in Recommendation:**
   - SVD helps us discover hidden patterns in user-item interactions.
   - It allows us to represent users and items in a lower-dimensional space of latent factors.
   - We can then predict how much a user would like a particular item based on their preferences and the item's characteristics.
   - This enables us to make personalized recommendations for users by suggesting items that they are likely to enjoy but have not yet interacted with.

5. **Using SVD in Recommendation Systems:**
   - In practice, we collect user-item interaction data (e.g., ratings, likes, purchases) to build a user-item matrix.
   - We apply SVD to decompose this matrix and learn the latent factors.
   - We then use the learned factors to predict ratings for unrated items and generate recommendations for users.
   - Finally, we evaluate the performance of our recommendation system using metrics like Root Mean Squared Error (RMSE) to ensure that the predictions are accurate.

In summary, SVD is a powerful technique used in recommendation systems to uncover hidden patterns in user-item interactions and make personalized recommendations for users. It helps in understanding user preferences and suggesting relevant items, enhancing the user experience and driving engagement.

```python
from surprise import Dataset, Reader
from surprise import SVD
from surprise.model_selection import train_test_split
from surprise import accuracy

# Assuming your DataFrame is called df
# You need to create a Dataset object from your DataFrame
reader = Reader(rating_scale=(1, 100))
data = Dataset.load_from_df(track_counts[['user_id', 'track', 'track_count']], reader)

# Split the dataset into train and test sets
trainset, testset = train_test_split(data, test_size=0.2, random_state=42)

# Initialize the SVD algorithm
algo = SVD()

# Train the algorithm on the trainset
algo.fit(trainset)

# Make predictions on the testset
predictions = algo.test(testset)

# Evaluate the accuracy of the model
accuracy.rmse(predictions)
```

```python
# Function to get top-N recommendations for a user
def get_top_n_recommendations(user_id, n=20):
    # Get all items the user has not rated
    items_to_rate = [item for item in trainset.all_items() if item not in trainset.ur[user_id]]

    # Predict ratings for these items
    item_ratings = [(item, algo.predict(user_id, item).est) for item in items_to_rate]

    # Sort the predictions by estimated rating
    item_ratings.sort(key=lambda x: x[1], reverse=True)

    # Get top-N recommendations
    top_n_recommendations = item_ratings[:n]

    # Convert item IDs to item names
    top_n_recommendations_with_names = [(trainset.to_raw_iid(item_id), rating) for item_id, rating in top_n_recommendations]

    return top_n_recommendations_with_names
```

# Let's Create a Pipeline .

# 1. data_collection.py

```python
import pandas as pd

def read_csv(file_path):
    df = pd.read_csv(file_path, on_bad_lines='skip')
    return df
```

# 2. data_transformation.py

```python
import re

def clean_data(df):
    # Rename columns
    column_rename_mapping = {
        ' "artistname"':'artist',
        ' "trackname"': 'track',
        ' "playlistname"': 'playlist'
    }
    df = df.rename(columns=column_rename_mapping)

    # Clean text values
    columns_to_clean = ['artist', 'track', 'playlist']
    for col_name in columns_to_clean:
        df[col_name] = df[col_name].astype(str)
        df[col_name] = df[col_name].apply(lambda x: re.sub(r'[^\w\s]', '', x))

    # Group by 'user_id' and 'track' and count occurrences
    track_counts = df.groupby(['user_id', 'track']).size().reset_index(name='track_count')

    return track_counts
```

# 3. model_training.py

```python
!pip install scikit-surprise

from surprise import Dataset, Reader
from surprise import SVD
from surprise.model_selection import train_test_split
from surprise import accuracy

def build_and_train_model(track_counts):
    reader = Reader(rating_scale=(1, 100))
    data = Dataset.load_from_df(track_counts[['user_id', 'track', 'track_count']], reader)
    trainset, testset = train_test_split(data, test_size=0.2, random_state=42)

    algo = SVD()
    algo.fit(trainset)

    predictions = algo.test(testset)
    rmse = accuracy.rmse(predictions)

    return algo, rmse
```

# 4. model_prediction.py

```python
def get_top_n_recommendations(algo, trainset, custom_user_id, n=20):
    items_to_rate = [item for item in trainset.all_items() if item not in trainset.ur[custom_user_id]]
    item_ratings = [(item, algo.predict(custom_user_id, item).est) for item in items_to_rate]
    item_ratings.sort(key=lambda x: x[1], reverse=True)
    top_n_recommendations = item_ratings[:n]
    top_n_recommendations_with_names = [(trainset.to_raw_iid(item_id), rating) for item_id, rating in top_n_recommendations]
    return top_n_recommendations_with_names
```

# Now, let's use these modules to create the final pipeline:

# 5. training_pipeline.py

```python
from data_collection import read_csv
from data_transformation import clean_data
from model_training import build_and_train_model
import joblib

def main(file_path, model_save_path):
    # Data Collection
    df = read_csv(file_path)

    # Data Transformation
    track_counts = clean_data(df)

    # Model Building and Training
    algo, rmse = build_and_train_model(track_counts)

    # Save the trained model to a file
    joblib.dump(algo, model_save_path)
    print("Model saved to:", model_save_path)

if __name__ == "__main__":
    # Define file path and model save path
    file_path = '/content/drive/MyDrive/Data Science Projects/ML Intern@CodeAlpha/Task_1_Music Recommendation System/dataset_2.csv'
    model_save_path = 'music_recommender_model.joblib'

    # Execute the main function
    main(file_path, model_save_path)
```

# 6. pred_pipeline.py

```python
import joblib
from model_prediction import get_top_n_recommendations

def predict_recommendations(model_load_path, custom_user_id):
    # Load the trained model from the joblib file
    loaded_algo = joblib.load(model_load_path)

    # Prediction using the loaded model
    recommendations = get_top_n_recommendations(loaded_algo, loaded_algo.trainset, custom_user_id)
    return recommendations

if __name__ == "__main__":
    # Define model load path
    model_load_path = 'music_recommender_model.joblib'

    # Input custom user ID
    custom_user_id = input("Enter the custom user ID: ")

    # Predict recommendations using the loaded model
    recommendations = predict_recommendations(model_load_path, custom_user_id)
    print("Top Recommendations using loaded model:")
    print(recommendations)
```
